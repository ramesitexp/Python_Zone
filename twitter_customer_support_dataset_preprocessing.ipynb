```python
# html,url attention
import re
def remove_html_tags(text):
    pattern=re.compile(r'<.*?>')
    return pattern.sub('',str(text))
```

```python
data["text"]=data['text'].apply(remove_html_tags)
```

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",str(text))
```

```python
data['text']=data['text'].apply(remove_url)
```

```python
data.head(10)
```

```python
#punctuations
import string
exclude=string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
data['text']=data['text'].apply(remove_punc)
```

```python
data.head(10)
```

```python
#Slang and abbreviations

chat_words={
   " AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LAMO":"Laugh My A.. Off",
"FYI":"For your information"
}
```

```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
data['text']=data['text'].apply(chat_conversion)
```

```python
#spelling correction
import textblob
from textblob import TextBlob
def remove_spell_err(text):
    txtblob=TextBlob(text)
    return txtblob
```

```python
data['text']=data['text'].apply(remove_spell_err)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
data.head()
```

```python
def remove_stopwords(text):
    new_text=[]

    for word in text.split():
        if str(word) in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())


    return " ".join(str(new_text)).replace("   ","")
```

```python
data['text']=data['text'].apply(remove_stopwords)
```

```python
import re
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', str(text))
    return clean_text
```

```python
data['text']=data['text'].apply(remove_emojis_manually)
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```
